### C18_SpectogramColors.ipynb
### Bu adımda kick ve snare seslerinden 3'er örnek kullanılarak bu sinyallerin spektrogram görüntüleri incelendi. Spektogram grafiğini oluşturmak için librosa modülü kullanıldı.
### Bu aşamada kullanılan kodlar için kaynaklar:
### https://librosa.org/doc/latest/index.html
### https://medium.com/datarunner/librosa-9729c09ecf7a

```python
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa.display
%matplotlib inline
```

# ilk kick örneği:

```python
ses = "C18_kick-22in.wav"
x,sr = librosa.load(ses)
print("x type:",type(x),"sr type",type(sr))
print(x.shape,sr)

#-- spektogram --#
X = librosa.stft(x) #stft -> Short-time Fourier transform
Xdb=librosa.amplitude_to_db(abs(X)) #Genlikten Desibel değerine

plt.figure(figsize=(20,8))
librosa.display.specshow(Xdb,sr=sr,x_axis="time",y_axis="hz")
plt.colorbar()

#-- sesi oynatma --#
ipd.Audio(ses)
ipd.Audio(x,rate=sr) #Bu şekilde de oynatılabiliyor.
```
