### C18_ZeroCrossingRate.ipynb
### Bu adımda kick ve snare seslerinden 3'er örnek kullanılarak bu sinyallerin zero crossing rate grafikleri incelendi. Zero crossing rate grafiğini oluşturmak ve bu değere ulaşmak için librosa modülü kullanıldı.
### Bu aşamada kullanılan kodlar için kaynaklar:
### https://librosa.org/doc/latest/index.html
### https://medium.com/datarunner/librosa-9729c09ecf7a

```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
from IPython.lib.display import Audio

%matplotlib inline
```

```python
ses1 = "C18_kick-22in.wav"
x1,sr = librosa.load(ses1)
print("x type:",type(x1),"sr type",type(sr))
print(x1.shape,sr)

zero_crossing_of_x1 = librosa.zero_crossings(x1)
print(sum(zero_crossing_of_x1)) #Toplam sıfır geçişi sayısı

plt.figure()
plt.plot(x1[1000:3000])

plt.title("kick-22in örneği için zero crossing rate grafiği")
plt.grid()

print('x1[n] işareti için:')
display(Audio(x1, rate=sr))
```
